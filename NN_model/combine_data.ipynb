```python
import numpy as np
import pandas as pd

b_data = pd.read_excel("data/basic_data.xlsx")
a_data = pd.read_excel("data/advanced_data.xlsx")
# Rename the merge columsn
b_data.rename(columns={"Season Start":"year", "Player Name":"player"}, inplace=True)
data = b_data.merge(a_data, on=["year", "player"])
data = data[data.year >= 2000]
salary_caps = pd.read_csv("data/salary_cap.csv", header=None).iloc[:, :2]
salary_caps.rename(columns={0:"year", 1:"salary_cap"}, inplace=True)
salary_caps["year"] = salary_caps["year"].apply(lambda val: int(str(val)[:4]))
salary_caps["salary_cap"] = salary_caps["salary_cap"].apply(lambda val: 
                                                            float(val.replace("$", "").replace(",","")))
data = data.merge(salary_caps, on=["year"])
data.rename(columns={"Player Salary in $":"salary"}, inplace=True)
# Modify the target variable to make it a percentage
target = "salary"
data = data[pd.isna(data[target])==False]
data[target] = data.apply(lambda val: val[target]/val["salary_cap"], axis=1)
```

```python
# Save the data as a pickle
data.to_pickle("data/main.df")
```
